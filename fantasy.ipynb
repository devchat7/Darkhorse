```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage
url = 'https://fftoolbox.fulltimefantasy.com/football/strength_of_schedule.cfm?type=f&sortby=K'

# Fetch the HTML content from the URL with SSL verification disabled
response = requests.get(url, verify=False)
if response.status_code == 200:
    html_content = response.text

    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table rows
    rows = soup.find_all('tr', class_='c')

    # List to hold the extracted data
    data = []

    for row in rows:
        # Extract the text from each td element in the row
        row_data = [td.get_text(strip=True) for td in row.find_all('td')]
        data.append(row_data)

    # Define column names
    columns = ['TEAM', 'QB', 'RB', 'WR', 'TE', 'K', 'DEF', 'DL', 'LB', 'DB', 'PR', 'KR']

    # Create a DataFrame
    ScheduleDifficulty = pd.DataFrame(data, columns=columns)

    # Create a DataFrame
    ScheduleDifficulty = pd.DataFrame(data, columns=columns)

    # Replace 'LVR' with 'LV' in the TEAM column
    ScheduleDifficulty['TEAM'] = ScheduleDifficulty['TEAM'].replace('LVR', 'LV')

    ScheduleDifficulty['TEAM'] = ScheduleDifficulty['TEAM'].replace('LAR', 'LA')

    # Select only the relevant columns
    ScheduleDifficulty = ScheduleDifficulty[["TEAM", "QB", "RB", "WR", "TE", "K", "DEF"]]

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

print(ScheduleDifficulty)
```

```python
def add_games_played_column(weekly_data):
    weekly_data['GP'] = weekly_data.groupby('player_display_name')['week'].transform('count')
    return weekly_data

data_2023 = nfl.import_weekly_data([2023])
weekly_data_2023 = data_2023.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2023 = add_games_played_column(weekly_data_2023)

data_2022 = nfl.import_weekly_data([2022])
weekly_data_2022 = data_2022.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2022 = add_games_played_column(weekly_data_2022)

data_2021 = nfl.import_weekly_data([2021])
weekly_data_2021 = data_2021.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2021 = add_games_played_column(weekly_data_2021)

data_2020 = nfl.import_weekly_data([2020])
weekly_data_2020 = data_2020.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2020 = add_games_played_column(weekly_data_2020)

data_2019 = nfl.import_weekly_data([2019])
weekly_data_2019 = data_2019.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2019 = add_games_played_column(weekly_data_2019)

data_2018 = nfl.import_weekly_data([2018])
weekly_data_2018 = data_2018.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2018 = add_games_played_column(weekly_data_2018)

# Display the columns of the 2018 weekly data to verify the addition of the "GP" column
weekly_data_2018

```

```python
# Function to merge recent team into seasonal data
def merge_recent_team(seasonal_data, weekly_data):
    grouped_data = weekly_data.groupby(['player_display_name']).agg({
        'player_id': 'first',
        'recent_team': 'last',
        'GP': 'first',
        'position': 'first'  # Get the most recent team
    }).reset_index()
    merged_data = pd.merge(grouped_data, seasonal_data, on='player_id', how='left')
    return merged_data

# Import and merge seasonal data for each year
season_data_2023 = nfl.import_seasonal_data([2023])
season_data_2023 = merge_recent_team(season_data_2023, weekly_data_2023)
print(season_data_2023)

season_data_2022 = nfl.import_seasonal_data([2022])
season_data_2022 = merge_recent_team(season_data_2022, weekly_data_2022)
print(season_data_2022)

season_data_2021 = nfl.import_seasonal_data([2021])
season_data_2021 = merge_recent_team(season_data_2021, weekly_data_2021)
print(season_data_2021)

season_data_2020 = nfl.import_seasonal_data([2020])
season_data_2020 = merge_recent_team(season_data_2020, weekly_data_2020)
print(season_data_2020)

season_data_2019 = nfl.import_seasonal_data([2019])
season_data_2019 = merge_recent_team(season_data_2019, weekly_data_2019)
print(season_data_2019)

season_data_2018 = nfl.import_seasonal_data([2018])
season_data_2018 = merge_recent_team(season_data_2018, weekly_data_2018)
print(season_data_2018)
season_data_2018.columns
```

```python
import pandas as pd

# Assuming you have the seasonal data loaded in variables named season_data_2018 to season_data_2023
historical_data = pd.concat([season_data_2018, season_data_2019, season_data_2020, season_data_2021, season_data_2022])

# Convert team names in both dataframes to uppercase for consistency
historical_data['recent_team'] = historical_data['recent_team'].str.upper()
ScheduleDifficulty['TEAM'] = ScheduleDifficulty['TEAM'].str.upper()

# Merge the historical data with the strength of schedule data
merged_data = pd.merge(historical_data, ScheduleDifficulty, left_on='recent_team', right_on='TEAM', how='left')

merged_data['avg_fantasy_points'] = merged_data['fantasy_points'] / merged_data['GP']

print(merged_data.columns)

# Remove rows with NaN values
merged_data = merged_data.dropna()

# Print the cleaned data
print("Cleaned data without NaN values:")
print(merged_data)
```

```python
import pandas as pd

# Assuming merged_data is your DataFrame
numeric_columns = merged_data.select_dtypes(include=['number']).columns
print("Numeric columns in merged_data:")
print(numeric_columns)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Select numeric features and target variable
features = merged_data[numeric_columns].drop(columns=['avg_fantasy_points'])  # Exclude the target variable
player_info = merged_data[['player_display_name', 'player_id']]
target = merged_data['avg_fantasy_points']

# Handle any missing values
features = features.fillna(0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test, player_train, player_test = train_test_split(features, target, player_info, test_size=0.2, random_state=42)

# Train the Decision Tree model
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train, y_train)

# Extract feature importances
feature_importances = tree_model.feature_importances_
feature_names = features.columns

# Create a DataFrame for feature importances
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print("Feature importances:")
print(importance_df)
```

```python
# Select top features based on importance
top_features = importance_df['Feature'].tolist()
print("Top features selected based on importance:")
print(top_features)

# Select the top features from the data
top_feature_data = features[top_features]

# Split the data into training and testing sets again with top features
X_train, X_test, y_train, y_test, player_train, player_test = train_test_split(
    top_feature_data, target, player_info, test_size=0.2, random_state=42
)

# Train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")

# Combine predictions with player information
results = pd.DataFrame({
    'player_display_name': player_test['player_display_name'],
    'player_id': player_test['player_id'],
    'actual_fantasy_points': y_test,
    'predicted_fantasy_points': predictions
})

# Display some predictions
print(results.head(10))
```
